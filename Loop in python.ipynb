## Max and Minimum Problem


```python
print(" Enter 3 value x,y, z")
inp = input()
number = inp.split()

x = int(number[0])
y = int(number[1])
z = int(number[2])

minimum = x
maximum = x

# check minimum
if y < minimum:
    minimum = y
if z < minimum:
    minimum = z

# check maximum
if y > maximum:
    maximum = y
if z > maximum:
    maximum = z

print(minimum, maximum)
```

```python
# another way
inp = input("Enter 3 values: ")

# replace comma with space, then split
number = inp.replace(',', ' ').split()

if len(number) != 3:
    print("Please enter exactly 3 numbers")
else:
    x, y, z = map(int, number)
    print(min(x, y, z), max(x, y, z))
```

```python
#short cut
x, y, z = map(int, input().replace(',', ' ').split())
print(min(x, y, z), max(x, y, z))
```

## Even Odd Positive and Negetive


```python
n = int(input("Enter how many numbers: "))
inp = input("Enter numbers: ")

numbers = inp.split()

positive = 0
negative = 0
odd = 0
even = 0   # ✅ missing ছিল

for i in range(n):
    x = int(numbers[i])

    # positive or negative
    if x > 0:
        positive += 1
    elif x < 0:
        negative += 1

    # odd or even
    if x % 2 != 0:
        odd += 1
    else:
        even += 1

print("Even:", even)
print("Odd:", odd)
print("Positive:", positive)
print("Negative:", negative)
```

```python
## Anoter one
numbers = list(map(int, input("Enter numbers: ").split()))

positive = negative = odd = even = 0

for x in numbers:
    if x > 0:
        positive += 1
    elif x < 0:
        negative += 1

    if x % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even:", even)
print("Odd:", odd)
print("Positive:", positive)
print("Negative:", negative)
```

## Digits 

```python
t = int(input("Enter number of test cases: "))

for i in range(t):
    number = int(input("Enter a number: "))

    if number == 0:
        print(0)
        continue

    while number > 0:
        print(number % 10, end="")
        number //= 10
    print()
```

```python
#Short cut
t = int(input())

for i in range(t):
    num = input()
    print(num[::-1])
    
```
